```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the CSV file
teamrating = pd.read_csv('/Users/stevepham/Downloads/23-24TeamRatings.csv')
```

```python
teams_to_plot = ['Boston Celtics', 'New York Knicks', 'Milwaukee Bucks', 
                 'Cleveland Cavaliers', 'Orlando Magic']  

# Filter the DataFrame
filtered_teamrating = teamrating[teamrating['Team'].isin(teams_to_plot)]

# Plot W/L%
plt.figure(figsize=(10,6))
plt.bar(filtered_teamrating['Team'], filtered_teamrating['W/L%'])
plt.title('Win/Loss % of Eastern Conference Top 5 Teams (2023-2024 Season)')
plt.ylabel('W/L%')
plt.xlabel('Team')
plt.ylim(0, 1) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
teams = filtered_teamrating['Team']
ortg_a = filtered_teamrating['ORtg/A']
drtg_a = filtered_teamrating['DRtg/A']

# League-average ORtg/DRtg
league_avg_ortg = teamrating['ORtg/A'].mean()
league_avg_drtg = teamrating['DRtg/A'].mean()
print(f'Average Offensive Rating: {league_avg_ortg:.2f}')
print(f'Average Defensive Rating: {league_avg_drtg:.2f}')

# Set positions for grouped bars (used AI to help graph two variables
# On the same axes)
x = np.arange(len(teams))  # positions on x-axis
width = 0.35  # width of the bars

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, ortg_a, width, label='Adjusted Offensive Rating (ORtg)')
plt.bar(x + width/2, drtg_a, width, label='Adjusted Defensive Rating (DRtg)')

# League Average Showings
# Plot league averages as dotted lines
plt.axhline(league_avg_ortg, color='red', linestyle='--', linewidth=1.5, label='League Avg ORtg')
plt.axhline(league_avg_drtg + 0.1, color='black', linestyle='--', linewidth=1.5, label='League Avg DRtg')

# Legend, axes, etc.
plt.xlabel('Team')
plt.ylabel('Rating')
plt.title('Adjusted Offensive and Defensive Ratings (2023-2024)')
plt.xticks(x, teams)
plt.ylim(95, max(max(ortg_a), max(drtg_a)) + 5)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
