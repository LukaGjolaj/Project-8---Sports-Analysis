```python
import pandas as pd
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = pd.read_csv("Cleaned_TeamAdvancedStats.csv")
X = df[['SRS', 'MOV', 'SOS', 'Age', 'ORtg', 'DRtg','TS%']]
X = sm.add_constant(X)
y = df['W']
# Fit the model
model = sm.OLS(y, X).fit()
# Summary
print(model.summary())
```

```python
y_pred = model.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred, color='blue', alpha=0.6, label="Predicted vs Actual")
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--', linewidth=2, label='Perfect Prediction')
plt.xlabel("Actual Wins")
plt.ylabel("Predicted Wins")
plt.title("Statsmodels: Actual vs Predicted Wins")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("Cleaned_TeamAdvancedStats.csv")
X = df[['ORtg']]  # Independent variable
y = df['W']      # Dependent variable

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Predict for the line
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_pred = model.predict(x_range)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(x_range, y_pred, color='red', linewidth=2, label='Linear Regression Line')
plt.xlabel("ORTG")
plt.ylabel("Wins")
plt.title("Linear Regression: Wins vs. ORTG")
plt.legend()
plt.grid(True)
plt.show()
```
