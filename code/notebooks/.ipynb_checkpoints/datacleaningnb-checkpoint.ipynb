```python
import pandas as pd
```

```python
teamrating = pd.read_csv('23-24TeamRatings.csv')
teamrating.head()
```

```python
teamadvanced = pd.read_csv('23-24TeamAdvanceStats.csv')
teamadvanced = teamadvanced.drop(columns=['Unnamed: 22', 'Unnamed: 27'])
teamadvanced['Team'] = teamadvanced['Team'].str.replace('*', '', regex=False).str.strip()
teamadvanced.rename(columns={
    'eFG%.1': 'eFG%',
    'TOV%.1': 'TOV%',
    'FT/FGA.1': 'Opponent FT/FGA'
}, inplace=True)
teamadvanced = teamadvanced[teamadvanced['Team'] != 'League Average']
teamadvanced.head(31)
teamadvanced.dtypes

#teamadvanced.to_csv('Cleaned_TeamAdvancedStats.csv', index=False)
```

```python
shootingtypes= pd.read_csv('23-24ShootingTypePerTeamStats.csv')
shootingtypes = shootingtypes.drop(columns=[ 'Unnamed: 13',
    'Unnamed: 6', 'Unnamed: 26', 'Unnamed: 29', 'Unnamed: 32', 'Unnamed: 23','Unnamed: 20'
])
shootingtypes['Team'] = shootingtypes['Team'].str.replace('*', '', regex=False).str.strip()
shootingtypes.rename(columns={
    '10-Mar': '3-10',
    '16-Oct': '10-16',
    '2P.1':'2PFGM%',
    '0-3.1': '0-3FGM%',
    '10-Mar.1':'3-10FGM%',
    'Att.':'Heaves Attempt',
    '16-Oct.1':'10-16PFGM%',
    '16-3P.1':'16-3PFGM%',
    '3P.1':'3PFGM%',
    '2P.2':'%Astd2P',
    '3P.2':'%Asted3P',
    '%FGA':'%FGDunks',
    'Md.':'TotalDunksMade',
    '%FGA.1':'%FGLayups',
    'Md..1':'TotalLayups',
    '3%PA':'%Corner3PA',
    '3P%':'3P%Corner',
    'Md..2': 'Heaves Made'
}, inplace=True)
shootingtypes.dtypes
#shootingtypes.to_csv('Cleaned_ShootingTypeStats.csv', index=False)
```

```python
teamstats = pd.read_csv('23-24TeamPerGameStats.csv')
teamstats['Team'] = teamstats['Team'].str.replace('*', '', regex=False).str.strip()
teamstats = teamstats.iloc[:-1]
teamstats.head(32)
teamstats.to_csv('Cleaned_TeamStats.csv', index=False)
```
