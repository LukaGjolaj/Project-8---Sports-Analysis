```python
player_info = pd.read_csv('../../data/rawdata/PlayerInfo.csv')



def height_wingspan_to_inches(measurement_str):
    """Converts a string like "F'I.XX\"" (e.g., "6'11.25\"") to total inches."""
    # Handle missing data (like NaN in the Team column showing up if parsing was imperfect)
    if pd.isna(measurement_str) or not isinstance(measurement_str, str):
        return np.nan # Return Not a Number for missing values
    
    try:
        # Split the string at the foot marker (')
        parts = measurement_str.split("'")
        feet = int(parts[0])
        
        # Remove the inch marker (") and convert the inches part to a float
        inches = float(parts[1].replace('"', ''))
        
        # Calculate total inches
        total_inches = (feet * 12) + inches
        return total_inches
    except (IndexError, ValueError):
        # Handle cases where the string format might be unexpected
        print(f"Warning: Could not parse '{measurement_str}'. Returning NaN.")
        return np.nan

# Apply the conversion function to the 'Height' column
player_info['Height_Inches'] = player_info['Height'].apply(height_wingspan_to_inches)

# Apply the conversion function to the 'Wingspan' column
player_info['Wingspan_Inches'] = player_info['Wingspan'].apply(height_wingspan_to_inches)

columns = ['Player', 'Team', 'Position', 'Height_Inches', 'Wingspan_Inches']

player_info_cleaned = player_info[columns]

player_info_cleaned.to_csv('../../data/cleandata/Cleaned_PlayerInfo.csv')
```

```python
cleaned_player_length = pd.read_csv('../../data/cleandata/Cleaned_PlayerInfo.csv')
cleaned_player_length

raw_player_minutes = pd.read_csv('../../data/rawdata/PlayerMinutes.csv')
raw_player_minutes

player_merged_data = pd.merge(
        cleaned_player_length,
        raw_player_minutes[['PLAYER_NAME', 'TEAM_ABBREVIATION', 'MIN']], # Select only the needed columns from the minutes data
        left_on=['Player', 'Team'],           # Columns from cleaned_player_length
        right_on=['PLAYER_NAME', 'TEAM_ABBREVIATION'], # Columns from raw_player_minutes
        how='left'                             # Keep all rows from the left dataframe
    )
player_merged_data = player_merged_data.drop(columns=['PLAYER_NAME', 'TEAM_ABBREVIATION'])

columns = ['Player', 'Team', 'Position', 'Height_Inches', 'Wingspan_Inches', 'MIN']
player_merged_data = player_merged_data[columns]
player_merged_data.dropna(subset=['MIN'], inplace=True) 
player_merged_data = player_merged_data[player_merged_data['MIN'] > 0].copy()

player_merged_data.to_csv('../../data/cleandata/Cleaned_PlayerMinutesLength', index=False)
```

```python
player_merged_data['H_x_MP'] = player_merged_data['Height_Inches'] * player_merged_data['MIN']

team_grouped = player_merged_data.groupby('Team') 

team_agg_h = team_grouped.agg(
    Total_MP_H=('MIN', 'sum'),
    Sum_H_x_MP=('H_x_MP', 'sum')
)
team_agg_h['Avg_Height_Weighted'] = team_agg_h['Sum_H_x_MP'] / team_agg_h['Total_MP_H']


player_data_ws = player_merged_data.dropna(subset=['Wingspan_Inches']).copy()

player_data_ws['WS_x_MP'] = player_data_ws['Wingspan_Inches'] * player_data_ws['MIN']

team_grouped_ws = player_data_ws.groupby('Team') # Group again on the filtered data

team_agg_ws = team_grouped_ws.agg(
    Total_MP_WS=('MIN', 'sum'),
    Sum_WS_x_MP=('WS_x_MP', 'sum')
)
team_agg_ws['Avg_Wingspan_Weighted'] = team_agg_ws['Sum_WS_x_MP'] / team_agg_ws['Total_MP_WS']

# --- Combine height and wingspan averages ---
team_weighted_stats = pd.merge(
    team_agg_h[['Avg_Height_Weighted']], 
    team_agg_ws[['Avg_Wingspan_Weighted']], 
    left_index=True,  # Merge on Team index from groupby
    right_index=True, 
    how='left'         # Keep all teams, even if they lack wingspan data (will have NaN)
)

team_weighted_stats.info()
```
