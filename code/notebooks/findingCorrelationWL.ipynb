```python
import pandas as pd
df = pd.read_csv("Project-8---Sports-Analysis/data/cleandata/23-24TeamRatings.csv")
df.head()
```

```python
df = df.drop(['Rk', 'Team', 'Conf', 'Div', 'W', 'L'], axis=1)
df.head()
```

```python
corr_matrix = df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
corr_matrix_lower = corr_matrix.mask(mask)
sns.heatmap(corr_matrix_lower, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Matrix', fontsize=14)
plt.show()
```
