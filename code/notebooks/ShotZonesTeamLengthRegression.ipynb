```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/cleandata/Cleaned_2024-2025TeamZonesLength.csv', index_col=False)
df
```

```python
predictor_cols = [
    'Avg_Wingspan_Weighted', 
    'RestrictedArea_FGA', 
    'MidRange_FGA', 
    '3_FGA'
]
target_col = 'W'
```

```python
# Create X and y
X = df[predictor_cols]
y = df[target_col]

# Verify shapes
print("\nShape of Predictors (X):", X.shape)
print("Shape of Target (y):", y.shape)
```

```python
import statsmodels.api as sm

# Add a constant (intercept) to the predictors
# OLS requires this to be done manually in statsmodels
X_with_const = sm.add_constant(X) 

# Create the OLS (Ordinary Least Squares) model
model = sm.OLS(y, X_with_const)

# Fit the model to the data
results = model.fit()

# Print the detailed summary report
print("\n--- Regression Results ---")
print(results.summary())
```
